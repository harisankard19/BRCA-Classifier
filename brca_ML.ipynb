```python
import pandas as pd
import numpy as np
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r"D:\Projects\brca\data\fineneedleaspirationData.csv")
df.head()
```

```python
df.shape
```

```python
df["diagnosis"].value_counts()
```

```python
#Data is already clean and filled with no missing values.
```

```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
df.head() #mapped properly
```

```python
df = df.drop(columns=['id'])
```

```python
df = df.drop(columns=['Unnamed: 32'])
```

```python
#here, we can see clear distinction between the two cases
```

```python
#INPUT FEATURES (X) and TARGET (y) split

X = df.drop(columns = "diagnosis", axis = 1)
y = df["diagnosis"]
```

```python
X.head(3)
```

```python
y.head(3)
```

```python
print(X.shape, y.shape)
```

```python
#TRAIN - TEST Splitting using sklearn

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state = 2) #20 test, 80 train split

#random_state is to improve reproducibility
```

```python
print(X.shape, Xtrain.shape, Xtest.shape)
print(y.shape, ytrain.shape, ytest.shape)
```

```python
#MODEL TRAINING

LRmodel = LogisticRegression()
```

```python
#TRAIN LRmodel using Xtrain, ytrain

LRmodel.fit(Xtrain, ytrain) #finds the best fit automatically
```

```python
#MODEL EVALUATION using Accuracy Score Metric

#Predict accuracy for Training data:

Xtrain_prediction = LRmodel.predict(Xtrain) #predicts the ytrain for given Xtrain
train_data_accuracy = accuracy_score(ytrain, Xtrain_prediction)
```

```python
print("Accuracy on Training data = ", train_data_accuracy)
```

```python
#Predict accuracy for Training data:

Xtest_prediction = LRmodel.predict(Xtest) #predicts the ytrain for given Xtrain
test_data_accuracy = accuracy_score(ytest, Xtest_prediction)

print("Accuracy on Test data = ", test_data_accuracy)
```
